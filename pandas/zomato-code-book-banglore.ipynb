```python
x = pd.read_csv("/kaggle/input/zomato-bangalore-restaurants/zomato.csv")
print(x.shape)
x.head()
```

```python
x.dtypes
```

- How many restaurants are in Bangalore?
- How many unique restaurants are in Bangalore.

```python
print("total unique restaurants: ", x["name"].value_counts().shape, x["name"].unique().shape)
print("total restaurants: ", x["name"].shape)
```

- how many restaurants are there in each area in Bangalore?

```python
df_rest_location_count= x["location"].value_counts().reset_index()
df_rest_location_count.columns = ["location", "count"]
df_rest_location_count.head()
```

- How many different restaurants types are there in each area in Bangalore?

```python
df_loc_rest_type_count = x.groupby(["location", "rest_type"])["url"].count().reset_index()
df_loc_rest_type_count.columns = ["location", "rest_type", "count"]
df_loc_rest_type_count.head()
```

- Check rest types at "BTM" location

```python
df_loc_rest_type_count[df_loc_rest_type_count["location"] == "BTM"]
```

- restaurant types in each area count and percentage?

```python
df_loc_rest_type_count_pivot = df_loc_rest_type_count.pivot(index="location", columns="rest_type", values="count").fillna(0)
df_loc_rest_type_count_pivot.head()
```

```python
percent_df_loc_rest_type_count_pivot =  (df_loc_rest_type_count_pivot.T / df_loc_rest_type_count_pivot.T.sum()).T
percent_df_loc_rest_type_count_pivot.head()
```

- which area has the highest restaurants?

```python
df_rest_location_count.sort_values("count", ascending=False)
```

```python
print(df_loc_rest_type_count_pivot.min(axis=1))
```

```python
percent_df_loc_rest_type_count = percent_df_loc_rest_type_count_pivot.stack().reset_index()
percent_df_loc_rest_type_count.columns = ["location", "rest_type", "percentage"]
percent_df_loc_rest_type_count.head()
```

```python
df1 = percent_df_loc_rest_type_count[percent_df_loc_rest_type_count["percentage"] > 0.5].reset_index(drop=True)
df1.head(), df1.shape
```

- Add location count restaurants

```python
out = pd.merge(df1, df_rest_location_count, on="location", how="left")
out.head()
```
